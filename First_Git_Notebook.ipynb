```python
from pyspark.sql.functions import *
```

```python
df = (
    spark.read.format("csv")
    .option("header", True)
    .option("InferSchema", True)
    .load("/Volumes/datamodeling/bronze/practice_read_write_demo/Order.csv")
)
```

```python
df = df.toDF(*[i.lower().replace(" ", "_") for i in df.columns])
```

```python
display(df.groupBy("order_priority").pivot("ship_mode").agg(sum(col("shipping_cost"))))
```

```python
display(df.groupBy(col('order_priority')).agg(collect_list(col('customer_segment'))))
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
data=[('vijay',1000),('pratap',2000),('raju',None),('kumar',None),('raghu',20000)]
schema=StructType([
       StructField('Name',StringType(),True),
       StructField('salary',IntegerType(),True)
])

emp_df=spark.createDataFrame(data,schema)
display(emp_df)
```

```python
avg_sal=emp_df.select(mean(col('salary'))).collect()[0][0]
```

```python
emp_df=emp_df.fillna({'salary':avg_sal})
```
